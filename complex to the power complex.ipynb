# Fixed base value

```python
v = []
K = []
for x in np.arange(1.0,20.0,0.05):
    for y in np.arange(1.0,20.0,0.05):
        z1 = complex(x,y)
        a = complex(1,1)
        for k in range(0,10):
            s = math.e**(z1*(math.log(abs(a),math.e)*(math.e**(j*(cmath.phase(a) + 2*k*math.pi)))))
            mag = abs(s)
            v.append((x,y,mag))
            K.append(k)
        
```

```python
xaxis = [i[0] for i in v]
yaxis = [i[1] for i in v]
zaxis = [i[2] for i in v]
plt.figure(figsize = (15,15))
ax = plt.axes(projection = '3d')
ax.plot3D(xaxis , yaxis , zaxis , c = 'blue',alpha = 0.7)
ax.set_zlabel("magnitude")
ax.set_title("Fixed base" , c = 'black')
ax.set_xlabel("real part of power")
ax.set_ylabel("imaginary part of power")
ax.view_init(-140,60)
```

# Varying base value

```python
x_log = []
y_log = []
mag_log = []
k_log = []
power = complex(2,2)
for real in np.arange(1,20,0.01):
    for imag in np.arange(1,20,0.01):
        base = complex(real,imag)
        for k in range(0,10):
            s = math.e**(power*(math.log(abs(base),math.e)*(math.e**(j*(cmath.phase(base) + 2*k*math.pi)))))
            x_log.append(real)
            y_log.append(imag)
            mag_log.append(abs(s))
            k_log.append(k)
```

```python
plt.figure(figsize = (14,14))
ax = plt.axes(projection = '3d')
ax.plot3D(x_log , y_log , mag_log , c = 'crimson' , alpha = 0.7)
ax.set_xlabel("real values of base")
ax.set_ylabel("imaginary values of base")
ax.set_zlabel("magnitude")
plt.title("VARYING BASE",c = "blue")
ax.view_init(-140,60)
```

```python
U,V,points = [] , [] , []
    # fig, axs = plt.subplots(2,5, figsize=(24, 14), facecolor='w', edgecolor='k')
    # fig.subplots_adjust(hspace = 1, wspace=1)
fig = plt.figure(figsize=plt.figaspect(0.5));
for A in np.arange(0,20,0.05):
    for B in np.arange(0,20,0.05):
        Z = complex(A,B)
        C = complex(1,1)
        u = ((math.e**(A*math.log(abs(C),math.e) - B*cmath.phase(C)))*(math.cos(A*np.arctan(10/10) + B*math.log(abs(C) , math.e))))
        U.append(u)
        v = ((math.e**(A*math.log(abs(C),math.e) - B*cmath.phase(C)))*(j*math.sin(A*np.arctan(10/10) + B*math.log(abs(C) , math.e))))
        V.append(abs(v))
        points.append((A,B))

xx , yy = [] , []
for i in points:
    xx.append(i[0])
    yy.append(i[1])

fig = plt.figure()
ax = plt.axes(projection = '3d')

ax.plot3D(xx , yy , U , c = 'mediumorchid')
ax.view_init(-150,60)
ax.set_title('trends in values of U')
ax.set_xlabel('values of x')
ax.set_ylabel('values of y')
ax.set_zlabel('values of u')
ax.view_init(-150,60)
plt.show()
fig = plt.figure()
axs = plt.axes(projection = '3d')
axs.plot3D(xx , yy , V , c = 'seagreen')
axs.set_title("values of V");
axs.set_xlabel('values of x');
axs.set_ylabel('values of y');
axs.set_zlabel('values of v');
axs.view_init(-150,60)
plt.show();
```
